```python
# start tcp server for streaming binance data

def start_tcp_server():
    import socket
    import time
    import json
    import random

    # create a socket object
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    # bind the socket to a local address and port
    s.bind(('localhost', 9999))

    # listen for incoming connections
    s.listen(1)
    print("TCP server is listening on port 9999...")

    while True:
        # accept a connection
        conn, addr = s.accept()
        print(f"Connection from {addr} has been established.")

        try:
            while True:
                # generate random data to simulate binance data
                data = {
                    "symbol": "BTCUSDT",
                    "price": round(random.uniform(30000, 40000), 2),
                    "timestamp": int(time.time() * 1000)
                }
                # send the data as a JSON string
                conn.sendall((json.dumps(data) + "\n").encode('utf-8'))
                time.sleep(1)  # send data every second
        except Exception as e:
            print(f"Connection from {addr} has been closed.")
            conn.close()
```
